# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df
sns.histplot(data=df, x='ConvertedComp', bins=60, kde=False)  # 'x' is the column name
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Values')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# SQL query
QUERY = """
SELECT Age
FROM master
"""

# Fetch data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Creating a box plot using Seaborn
sns.boxplot(data=df['Age'])  # Specify the column directly, no y parameter needed

plt.ylabel('Age')
plt.title('Box Plot of Age')

# Display the plot
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
sns.scatterplot(data=df, x='Age', y='WorkWeekHrs')

plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.title('Scatter Plot: Age vs Work Week Hours')


plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
sns.scatterplot(data=df, 
                x='WorkWeekHrs', 
                y='CodeRevHrs', 
                size='Age', 
                sizes=(20, 200), 
                legend='brief')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.title('Work Week Hours v Code Review Hours with Size by Age')

# Display the plot
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import numpy as np
```

```python
QUERY = """

select count(DatabaseDesireNextYear) as Count,DatabaseDesireNextYear

from DatabaseDesireNextYear 

group by DatabaseDesireNextYear 

order by count(DatabaseDesireNextYear) DESC LIMIT 5

"""

# Fetch data into a DataFrame
topdf = pd.read_sql_query(QUERY, conn)

# Creating a pie chart
plt.figure(figsize=(8, 8))
colors = sns.color_palette('Set3', len(topdf))  # Choose a color palette
plt.pie(topdf['Count'], labels=topdf['DatabaseDesireNextYear'],
        colors=colors, autopct='%1.1f%%', startangle=140, pctdistance=0.85,
        wedgeprops={'edgecolor': 'black', 'linewidth': 2})  # Add dark border to wedges

plt.gca().add_artist(plt.Circle((0, 0), 0.70, fc='lightgrey'))  # Light harmonizing background

plt.title('Top Desired Databases for Next Year')
plt.show()
```

```python
query = """
    SELECT LanguageDesireNextYear
    FROM LanguageDesireNextYear;
"""

# Execute the query and convert the result to a Pandas DataFrame
df = pd.read_sql_query(query, conn)

# Extract and count unique options
language_desire_counts = df['LanguageDesireNextYear'].value_counts()

# Get the top five options
top_five_languages = language_desire_counts.head(5)

# Create a bar plot
plt.figure(figsize=(10, 6))
top_five_languages.sort_values(ascending=True).plot(kind='barh', color='royalblue')
plt.xlabel('Frequency')
plt.ylabel('Programming Languages')
plt.title('Top Five Desired Programming Languages for Next Year')
plt.grid(True)
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35;

"""
df = pd.read_sql_query(QUERY, conn)
median_values = df.median()
ax = median_values.plot(kind='bar', stacked=True, figsize=(10, 6))
ax.set_xlabel('Code Review Hours')
ax.set_ylabel('Median Hours')
ax.set_title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xticks(rotation=0)
plt.show()
```

```python
# Define the SQL query
query = """
    SELECT Age, ConvertedComp
    FROM master
    WHERE Age BETWEEN 25 AND 30;
"""

# Execute the query and convert the result to a Pandas DataFrame
dfcorr = pd.read_sql_query(query, conn)



# Calculate the correlation
correlation = dfcorr['Age'].corr(dfcorr['ConvertedComp'])

print("Correlation between Age and ConvertedComp:", correlation)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60;

"""
df = pd.read_sql_query(QUERY, conn)
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['ConvertedComp'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median Converted Compensation for Age Group 45 to 60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master;

"""
df = pd.read_sql_query(QUERY, conn)
# Calculate the value counts for each category
main_branch_counts = df['MainBranch'].value_counts()

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='steelblue')
plt.xlabel('Frequency')
plt.ylabel('Main Branch')
plt.title('Frequency of Main Branch Categories')
plt.grid(True)
plt.show()
```
